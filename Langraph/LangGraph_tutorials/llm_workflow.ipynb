```python
import os
from typing import TypedDict

from dotenv import load_dotenv
from langchain_core.messages import HumanMessage
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langgraph.graph import END, START, StateGraph
```

```python
# Run cell 0 first. Instruct models: use ChatHuggingFace (chat_completion), not llm.invoke (text_generation).
load_dotenv()
token = os.getenv("HUGGINGFACEHUB_API_TOKEN") or os.getenv("HUGGINGFACE_API_KEY")
if not token:
    raise ValueError("Set HUGGINGFACE_API_KEY or HUGGINGFACEHUB_API_TOKEN in .env")
os.environ.setdefault("HUGGINGFACEHUB_API_TOKEN", token)

HF_MODEL_ID = "HuggingFaceTB/SmolLM2-1.7B-Instruct"

llm = HuggingFaceEndpoint(
    repo_id=HF_MODEL_ID,
    huggingfacehub_api_token=token,
    max_new_tokens=256,
    do_sample=False,
    temperature=0.2,
)
chat = ChatHuggingFace(llm=llm)

chat.invoke(
    [HumanMessage(content="In one short sentence, what does Hugging Face Hub provide?")]
)
```
